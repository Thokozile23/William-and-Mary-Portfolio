```python
# Define pre-pandemic and pandemic periods
pre_pandemic_period = retailer[retailer['WEEK_BEGIN_DATE'] < '2020-03-01'] # Use retailer DataFrame for pre-pandemic
pandemic_period = retailer[retailer['WEEK_BEGIN_DATE'] >= '2020-03-01'] # Use retailer DataFrame for pandemic

# Group by week and calculate total quantity sold
pre_pandemic_sales = pre_pandemic_period.groupby('WEEK_BEGIN_DATE')['SUB QTY'].sum()
pandemic_sales = pandemic_period.groupby('WEEK_BEGIN_DATE')['SUB QTY'].sum()

# Plot the sales trend
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(pre_pandemic_sales.index, pre_pandemic_sales, label='Pre-Pandemic')
plt.plot(pandemic_sales.index, pandemic_sales, label='Pandemic')
plt.title('Cheese Consumption Before and During the Pandemic')
plt.xlabel('Week')
plt.ylabel('Total Cheese Quantity Sold')
plt.legend()
plt.show()
```

```python
# Calculate price per unit (Revenue / Quantity)
retailer['price_per_unit'] = retailer['SUB REV'] / retailer['SUB QTY']

# Group by week to calculate average price per unit
avg_price_per_week = retailer.groupby('WEEK_BEGIN_DATE')['price_per_unit'].mean()

# Plot price trend
plt.figure(figsize=(10,6))
plt.plot(avg_price_per_week.index, avg_price_per_week, label='Average Cheese Price')
plt.title('Cheese Price Trends During the Pandemic')
plt.xlabel('Week')
plt.ylabel('Price per Unit')
plt.show()

# You can use external inflation data for comparison and correlation analysis
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pivot the data to create a matrix where rows are Zipcodes and columns are Store Names
shortages_pivot = ship.pivot_table(index='Zipcode', columns='STORE_NAME', values='QTY_NOT_SHP', aggfunc='sum', fill_value=0)

# Plot the heatmap
plt.figure(figsize=(12,8))
sns.heatmap(shortages_pivot, cmap='coolwarm', linewidths=0.5, annot=False)
plt.title('Heatmap of Cheese Shortages by Zipcode and Store')
plt.xlabel('Store Name')
plt.ylabel('Zipcode')
plt.show()
```

```python
# Analyze shortages
shortages = ship.groupby(['Zipcode', 'retailer_id'])['QTY_NOT_SHP'].sum()

# Plot shortages by retailer
shortages.plot(kind='bar', figsize=(12,6))
plt.title('Shortages by Retailer')
plt.xlabel('Retailer')
plt.ylabel('Total Quantity Not Shipped')
plt.show()

# Investigate shortages by zipcode (Demographic analysis can be done by joining with census data)
shortages_by_region = ship.groupby('Zipcode')['QTY_NOT_SHP'].sum()

# Plot shortages by region
shortages_by_region.plot(kind='bar', figsize=(12,6))
plt.title('Shortages by Zipcode')
plt.xlabel('Zipcode')
plt.ylabel('Total Quantity Not Shipped')
plt.show()
```
