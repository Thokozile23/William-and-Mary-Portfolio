```python

retailer['WEEK_BEGIN_DATE'] = pd.to_datetime(retailer['WEEK_BEGIN_DATE'])

# Define pre-pandemic and pandemic periods
pre_pandemic_period = retailer[(retailer['WEEK_BEGIN_DATE'] >= '2019-04-14') & (retailer['WEEK_BEGIN_DATE'] < '2020-12-04')] # Use retailer DataFrame for pre-pandemic
pandemic_period =  retailer[(retailer['WEEK_BEGIN_DATE'] >= '2020-12-04') & (retailer['WEEK_BEGIN_DATE'] <= '2023-05-23')]# Use retailer DataFrame for pandemic

# Group by week and calculate total quantity sold
pre_pandemic_sales = pre_pandemic_period.groupby('WEEK_BEGIN_DATE')['SUB QTY'].sum()
pandemic_sales = pandemic_period.groupby('WEEK_BEGIN_DATE')['SUB QTY'].sum()

# Plot the sales trend
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(pre_pandemic_sales.index, pre_pandemic_sales, color='green', label='Pre-Pandemic')
plt.plot(pandemic_sales.index, pandemic_sales, color='orange', label='Pandemic')

# Add labels, title, and legend
plt.title('Cheese Consumption Before and During the Pandemic')
plt.xlabel('Week', fontsize=12)
plt.ylabel('Total Cheese Quantity Sold', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)

# Format x-axis to show fewer ticks for clarity
# Specify custom x-axis ticks
custom_ticks = ['2019-04-14', '2020-02-08', '2020-12-04', '2021-09-30', '2022-07-27', '2023-05-23']
custom_ticks = pd.to_datetime(custom_ticks)  # Convert to datetime
plt.xticks(custom_ticks, rotation=45)


# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

#Analyzing cheese shortages
# Define the periods for pre-shortage, shortage, and post-shortage

pre_shortage = ship[(ship['WEEK_BEGIN_DATE'] >= '2019-04-14') & (ship['WEEK_BEGIN_DATE'] < '2020-12-04')]
shortage_period = ship[(ship['WEEK_BEGIN_DATE'] >= '2020-12-04') & (ship['WEEK_BEGIN_DATE'] <= '2023-05-23')]
post_shortage = ship[ship['WEEK_BEGIN_DATE'] >= '2023-05-23']

# Group by week and calculate total quantities for not shipped cheese
pre_shortage_qty = pre_shortage.groupby('WEEK_BEGIN_DATE')['QTY_NOT_SHP'].sum()
shortage_qty = shortage_period.groupby('WEEK_BEGIN_DATE')['QTY_NOT_SHP'].sum()
post_shortage_qty = post_shortage.groupby('WEEK_BEGIN_DATE')['QTY_NOT_SHP'].sum()

# Plot the trends
plt.figure(figsize=(12, 7))

# Fill areas for each period
plt.plot(pre_shortage_qty.index, pre_shortage_qty, color='green', label='Pre-Shortage')
plt.plot(shortage_qty.index, shortage_qty, color='orange', label='Labor Shortage')
plt.plot(post_shortage_qty.index, post_shortage_qty, color='green', label='Post-Shortage')

# Add labels, title, and legend
plt.title('Cheese Not Shipped Per Week', fontsize=14)
plt.xlabel('Week', fontsize=12)
plt.ylabel('Total Cheese Quantity Not Shipped (Thousands)', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)

# Format x-axis to show fewer ticks for clarity
custom_ticks = ['2019-04-14', '2020-02-08', '2020-12-04', '2021-09-30', '2022-07-27', '2023-05-23']
custom_ticks = pd.to_datetime(custom_ticks)  # Convert to datetime
plt.xticks(custom_ticks, rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure WEEK_BEGIN_DATE is in datetime format
supplier['WEEK_BEGIN_DATE'] = pd.to_datetime(supplier['WEEK_BEGIN_DATE'])

# Define periods for pre-shortage and shortage
pre_shortage = supplier[(supplier['WEEK_BEGIN_DATE'] >= '2019-04-14') & (supplier['WEEK_BEGIN_DATE'] < '2020-12-04')]
shortage = supplier[(supplier['WEEK_BEGIN_DATE'] >= '2020-12-04') & (supplier['WEEK_BEGIN_DATE'] <= '2023-05-23')]

# Group by week and calculate total quantities
pre_shortage_qty = pre_shortage.groupby('WEEK_BEGIN_DATE')['ARGO Quantity'].sum()
shortage_qty = shortage.groupby('WEEK_BEGIN_DATE')['ARGO Quantity'].sum()

# Plot the trends
plt.figure(figsize=(12, 7))

# Plot the data for each period
plt.plot(pre_shortage_qty.index, pre_shortage_qty, color='green', label='Pre-Shortage')
plt.plot(shortage_qty.index, shortage_qty, color='orange', label='Shortage')

# Add labels, title, and legend
plt.title('Cheese Ordered During the Pandemic', fontsize=14)
plt.xlabel('Week', fontsize=12)
plt.ylabel('Total Cheese Quantity Ordered (Thousands)', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)

# Specify custom x-axis ticks
custom_ticks = ['2019-04-14', '2020-02-08', '2020-12-04', '2021-09-30', '2022-07-27', '2023-05-23']
custom_ticks = pd.to_datetime(custom_ticks)  # Convert to datetime
plt.xticks(custom_ticks, rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```
